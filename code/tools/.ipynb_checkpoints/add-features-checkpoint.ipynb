```python
def getPolarity(text):
   return TextBlob(text).sentiment.polarity
```

```python
def getSubjectivity(text):
   return TextBlob(text).sentiment.subjectivity
```

```python
def getPartitionPolarity(text, partition_number, partition_index):
    start_index = len(text)//partition_number * (partition_index - 1)
    end_index = len(text)//partition_number * partition_index

    while start_index >= 0 and start_index < len(text) and text[start_index] != ' ':
        start_index = start_index + 1
    while end_index < len(text) and text[end_index] != ' ':
        end_index = end_index + 1
        
    return TextBlob(text[start_index:end_index]).sentiment.polarity
```

```python
def getPartitionSubjectivity(text, partition_number, partition_index):
    start_index = len(text)//partition_number * (partition_index - 1)
    end_index = len(text)//partition_number * partition_index

    while start_index >= 0 and start_index < len(text) and text[start_index] != ' ':
        start_index = start_index + 1
    while end_index < len(text) and text[end_index] != ' ':
        end_index = end_index + 1
        
    return TextBlob(text[start_index:end_index]).sentiment.subjectivity
```

```python
df['title_polarity'] = df['title'].apply(getPolarity)
```

```python
df['article_polarity'] = df['article'].apply(getPolarity)
```

```python
df['title_article_polarity_distance'] = df['article_polarity'] - df['title_polarity']
```

```python
df['article_polarity_first_third'] = df['article'].apply(getPartitionPolarity, args=(3,1))
```

```python
df['article_polarity_second_third'] = df['article'].apply(getPartitionPolarity, args=(3,2))
```

```python
df['article_polarity_third_third'] = df['article'].apply(getPartitionPolarity, args=(3,3))
```

```python
df['title_article_polarity_first_third_distance'] = df['article_polarity_first_third'] - df['title_polarity']
```

```python
df['title_article_polarity_second_third_distance'] = df['article_polarity_second_third'] - df['title_polarity']
```

```python
df['title_article_polarity_third_third_distance'] = df['article_polarity_third_third'] - df['title_polarity']
```

```python
df['title_subjectivity'] = df['title'].apply(getPolarity)
```

```python
df['article_subjectivity'] = df['article'].apply(getPolarity)
```

```python
df['title_article_subjectivity_distance'] = df['article_subjectivity'] - df['title_subjectivity']
```

```python
df['title_article_subjectivityy_first_third_distance'] = df['article_subjectivity_first_third'] - df['title_subjectivity']
```

```python
df['title_article_subjectivity_second_third_distance'] = df['article_subjectivity_second_third'] - df['title_subjectivity']
```
